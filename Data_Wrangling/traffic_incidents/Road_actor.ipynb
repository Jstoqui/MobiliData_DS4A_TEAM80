```python
peaton_fallecido = pd.DataFrame(df[(df['ESTADO']=='MUERTO') & (df['CONDICION']=='PEATON')].groupby(
    ['CODIGO_SINIESTRO','CONDICION'])['CONDICION'].count().reset_index(name='PEATON_FALLECIDO'))

peaton_herido = pd.DataFrame(df[(df['ESTADO']=='HERIDO') & (df['CONDICION']=='PEATON')].groupby(
    ['CODIGO_SINIESTRO','CONDICION'])['CONDICION'].count().reset_index(name='PEATON_HERIDO'))

pasajero_fallecido = pd.DataFrame(df[(df['ESTADO']=='MUERTO') & (df['CONDICION']=='PASAJERO')].groupby(
    ['CODIGO_SINIESTRO','CONDICION'])['CONDICION'].count().reset_index(name='PASAJERO_FALLECIDO'))

pasajero_herido = pd.DataFrame(df[(df['ESTADO']=='HERIDO') & (df['CONDICION']=='PASAJERO')].groupby(
    ['CODIGO_SINIESTRO','CONDICION'])['CONDICION'].count().reset_index(name='PASAJERO_HERIDO'))

conductor_fallecido = pd.DataFrame(df[(df['ESTADO']=='MUERTO') & (df['CONDICION']=='CONDUCTOR')].groupby(
    ['CODIGO_SINIESTRO','CONDICION'])['CONDICION'].count().reset_index(name='CONDUCTOR_FALLECIDO'))

conductor_herido = pd.DataFrame(df[(df['ESTADO']=='HERIDO') & (df['CONDICION']=='CONDUCTOR')].groupby(
    ['CODIGO_SINIESTRO','CONDICION'])['CONDICION'].count().reset_index(name='CONDUCTOR_HERIDO'))

motociclista_fallecido = pd.DataFrame(df[(df['ESTADO']=='MUERTO') & (df['CONDICION']=='MOTOCICLISTA')].groupby(
    ['CODIGO_SINIESTRO','CONDICION'])['CONDICION'].count().reset_index(name='MOTOCICLISTA_FALLECIDO'))

motociclista_herido = pd.DataFrame(df[(df['ESTADO']=='HERIDO') & (df['CONDICION']=='MOTOCICLISTA')].groupby(
    ['CODIGO_SINIESTRO','CONDICION'])['CONDICION'].count().reset_index(name='MOTOCICLISTA_HERIDO'))

ciclista_fallecido = pd.DataFrame(df[(df['ESTADO']=='MUERTO') & (df['CONDICION']=='CICLISTA')].groupby(
    ['CODIGO_SINIESTRO','CONDICION'])['CONDICION'].count().reset_index(name='CICLISTA_FALLECIDO'))

ciclista_herido = pd.DataFrame(df[(df['ESTADO']=='HERIDO') & (df['CONDICION']=='CICLISTA')].groupby(
    ['CODIGO_SINIESTRO','CONDICION'])['CONDICION'].count().reset_index(name='CICLISTA_HERIDO'))

del(peaton_fallecido['CONDICION'])
del(peaton_herido['CONDICION'])
del(pasajero_fallecido['CONDICION'])
del(pasajero_herido['CONDICION'])
del(conductor_fallecido['CONDICION'])
del(conductor_herido['CONDICION'])
del(motociclista_fallecido['CONDICION'])
del(motociclista_herido['CONDICION'])
del(ciclista_fallecido['CONDICION'])
del(ciclista_herido['CONDICION'])
```

```python
df_sexo = df[['CODIGO_SINIESTRO', 'SEXO']].copy()
```

```python
z = pd.DataFrame(df_sexo.groupby(
    ('CODIGO_SINIESTRO','SEXO'))['SEXO'].count().reset_index(name='CANTIDAD'))

z=z.pivot_table(
    index='CODIGO_SINIESTRO', columns='SEXO', values='CANTIDAD', fill_value = 0).reset_index()

z.rename_axis('', axis = 'columns', inplace=True) 
```

```python
actor_vial_sin_edad =  pd.merge(z, peaton_fallecido, how='outer' )
actor_vial_sin_edad = pd.merge(actor_vial_sin_edad, peaton_herido, how='outer')

actor_vial_sin_edad = pd.merge(actor_vial_sin_edad, pasajero_fallecido, how='outer')
actor_vial_sin_edad = pd.merge(actor_vial_sin_edad, pasajero_herido, how='outer')

actor_vial_sin_edad = pd.merge(actor_vial_sin_edad, conductor_fallecido, how='outer')
actor_vial_sin_edad = pd.merge(actor_vial_sin_edad, conductor_herido, how='outer')

actor_vial_sin_edad = pd.merge(actor_vial_sin_edad, ciclista_fallecido, how='outer')
actor_vial_sin_edad = pd.merge(actor_vial_sin_edad, ciclista_herido, how='outer')

actor_vial_sin_edad = pd.merge(actor_vial_sin_edad, motociclista_fallecido, how='outer')
actor_vial_sin_edad = pd.merge(actor_vial_sin_edad, motociclista_herido, how='outer')
```
